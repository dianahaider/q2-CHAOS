```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def pairplt (filenames, metadata=None):
    for i in range(len(filenames)-1):
        smpl = pd.merge(pd.read_csv(filenames[i]), pd.read_csv(filenames[i+1]), on = 'sample')
        if metadata:
            smpl.rename(columns={'sample':'sample-id'}, inplace=True)
            smpl = pd.merge(smpl,pd.read_csv(metadata), on = 'sample-id')
            sns.pairplot(smpl, hue = 'depth_code', vars = ['f2','f11'])
        else: return sns.pairplot(smpl, vars = ['f2','f11'])
    return
```
